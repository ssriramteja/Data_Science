 What is Feature Scaling?

 Feature scaling means converting data values into a similar range.





```python
import pandas as pd
```

```python
my_df = pd.DataFrame({"Height" : [1.98,1.77,1.76,1.80,1.64],
                      "Weight" : [99,81,70,86,82]})
my_df
```

# Standardisation

```python
from sklearn.preprocessing import StandardScaler

scale_standard = StandardScaler()
scale_standard.fit_transform(my_df)
my_df_standardised = pd.DataFrame(scale_standard.fit_transform(my_df), columns = my_df.columns)
my_df_standardised
```

```python
import matplotlib.pyplot as plt

# Example assumes you already have:
# my_df (original data)
# my_df_standardised (scaled data)

plt.figure(figsize=(14,6))

# -------- Left Plot (Unscaled) --------
plt.subplot(1,2,1)

plt.scatter(my_df.iloc[:,0], my_df.iloc[:,1], s=200)
plt.title("Unscaled")
plt.xlabel(my_df.columns[0])
plt.ylabel(my_df.columns[1])
plt.grid(True)

# -------- Right Plot (Standardized) --------
plt.subplot(1,2,2)

plt.scatter(my_df_standardised.iloc[:,0],
            my_df_standardised.iloc[:,1],
            s=200)

plt.title("Standardized")
plt.xlabel(my_df_standardised.columns[0])
plt.ylabel(my_df_standardised.columns[1])
plt.grid(True)

plt.tight_layout()
plt.show()
```

# normalisation

```python
from sklearn.preprocessing import MinMaxScaler

scale_norm = MinMaxScaler()
scale_norm.fit_transform(my_df)
my_df_normalised = pd.DataFrame(scale_norm.fit_transform(my_df), columns = my_df.columns)
my_df_normalised
```

```python
import matplotlib.pyplot as plt

#Creates a new plot window.
plt.figure(figsize=(6,6))

	#	scatter()  creates dot plot

""" First column of normalized dataframe.

iloc = index-based selection.
		:  all rows
		0  first column.
    s=200

Controls dot size.
		Larger value = bigger dots.
"""

plt.scatter(my_df_normalised.iloc[:,0],
            my_df_normalised.iloc[:,1],
            s=200)

plt.title("Normalized Data")


#Automatically uses column name.
plt.xlabel(my_df_normalised.columns[0])
plt.ylabel(my_df_normalised.columns[1])
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))

# -------- Left Plot (Standardized) --------
plt.subplot(1,2,1)

plt.scatter(my_df_standardised.iloc[:,0],
            my_df_standardised.iloc[:,1],
            s=200)

plt.title("Standardized")
plt.xlabel(my_df_standardised.columns[0])
plt.ylabel(my_df_standardised.columns[1])
plt.grid(True)

# -------- Right Plot (Normalized) --------
plt.subplot(1,2,2)

plt.scatter(my_df_normalised.iloc[:,0],
            my_df_normalised.iloc[:,1],
            s=200)

plt.title("Normalized")
plt.xlabel(my_df_normalised.columns[0])
plt.ylabel(my_df_normalised.columns[1])
plt.grid(True)

plt.tight_layout()
plt.show()
```
