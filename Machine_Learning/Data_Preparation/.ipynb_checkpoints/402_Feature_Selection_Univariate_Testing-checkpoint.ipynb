```python
import pandas as pd
```

```python
my_df = pd.read_csv("feature_selection_sample_data.csv")
my_df
```

# Regression Template

```python
from sklearn.feature_selection import SelectKBest, f_regression
```

```python
# we are droping the output column
x = my_df.drop(["output"], axis = 1)
# put the output column in Y
y = my_df["output"]
```

SelectKBest is a feature selection method from sklearn.

üëâ It selects the best features based on a scoring function.

Purpose:
	‚Ä¢	Remove less important features.
	‚Ä¢	Improve model performance.
	‚Ä¢	Reduce overfitting.
	‚Ä¢	Speed up training.

f_regression

Used when: Target variable is numeric (regression problem).


k="all"

This means:  Select ALL features.

```python
feature_selector = SelectKBest(f_regression, k = "all")
fit = feature_selector.fit(x,y)
```

f_values

F-statistic score for each feature.

What is p-value?

Probability that relationship happened by chance.

```python
p_values = pd.DataFrame(fit.pvalues_)
scores = pd.DataFrame(fit.scores_)
```

```python
input_variable_names = pd.DataFrame(x.columns)
input_variable_names
```

```python
summary_stats = pd.concat([input_variable_names,p_values,scores], axis = 1)
summary_stats.columns = ["input_variable","p_value","f_score"]
summary_stats.sort_values(by = "p_value", inplace = True)

summary_stats
```

```python
p_value_threshold = 0.05
score_threshold = 5
```

What this is doing

üëâ Selecting only the important features based on two conditions:

1Ô∏è‚É£ High F-score
2Ô∏è‚É£ Low p-value.

‚∏ª

‚≠ê Why are we performing this?

Because:

üëâ Not all features are useful for prediction.

Some features:
	‚Ä¢	Have weak relationship with target
	‚Ä¢	Add noise
	‚Ä¢	Reduce model accuracy.

So we filter.


```python
selected_variables = summary_stats.loc[(summary_stats["f_score"] >= score_threshold) & (summary_stats["p_value"] <= p_value_threshold)]
# This converts a column from a DataFrame into a Python list.
selected_variables = selected_variables["input_variable"].tolist()
X_new = x[selected_variables]
X_new
```

# Classification Template

```python
# the template is same but we perform chi2


from sklearn.feature_selection import SelectKBest, chi2

X = my_df.drop(["output"], axis = 1)
y = my_df["output"]

feature_selector = SelectKBest(chi2, k = "all")
fit = feature_selector.fit(X,y)

p_values = pd.DataFrame(fit.pvalues_)
scores = pd.DataFrame(fit.scores_)
input_variable_names = pd.DataFrame(X.columns)
summary_stats = pd.concat([input_variable_names,p_values,scores], axis = 1)
summary_stats.columns = ["input_variable","p_value","chi2_score"]
summary_stats.sort_values(by = "p_value", inplace = True)

p_value_threshold = 0.05
score_threshold = 5

selected_variables = summary_stats.loc[(summary_stats["chi2_score"] >= score_threshold) & (summary_stats["p_value"] <= p_value_threshold)]
selected_variables = selected_variables["input_variable"].tolist()
X_new1 = X[selected_variables]

X_new1
```
