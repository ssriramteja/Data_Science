                      "input2" : [29,41,44,47,50,53,56,59,66]})

vert=False

Controls orientation.

	•	vert=True (default) → vertical box plot
    
	•	vert=False → horizontal box plot

```python
my_df.plot(kind = "box", vert = False)
```

```python
outlier_columns = ["input1", "input2"]
```

# Boxplot approach

```python
for column in outlier_columns:
    #Calculate quartiles - These divide data into sections.
    lower_quartile = my_df[column].quantile(0.25)
    upper_quartile = my_df[column].quantile(0.75)
    # Calculate IQR -  middle spread of data.
    iqr = upper_quartile -lower_quartile
    iqr_extend =iqr * 1.5
    min_border = lower_quartile - iqr_extend
    max_border = upper_quartile + iqr_extend
    
    # Detect outliers -Select rows where value is: •	below minimum boundary OR •	above maximum boundary.
    outlier = my_df[(my_df[column] < min_border) | (my_df[column] > max_border)].index
#Shows how many were found.
    print(f"{len(outlier)} outliers detected in column {column}")
    
#Removes rows from dataset.

    my_df.drop(outlier, inplace =True)
```


# Standard Deviation approach

```python
my_df1 = pd.DataFrame({"input1" : [15,41,44,47,50,53,56,59,99],
                      "input2" : [29,41,44,47,50,53,56,59,66]})
```

```python
outlier_columns1 = ["input1", "input2"]
```

```python
for column1 in outlier_columns1:
    
    mean = my_df1[column1].mean()
    std_dev = my_df1[column1].std()

    min_border1 = mean - std_dev * 3
    max_border1 = mean + std_dev * 3

    outliers1 = my_df1[(my_df1[column1] < min_border1) | (my_df1[column1] > max_border1)].index

    print(f"{len(outliers1)} outliers detected in column {column1}")

    my_df.drop(outliers1, inplace= True)
```
