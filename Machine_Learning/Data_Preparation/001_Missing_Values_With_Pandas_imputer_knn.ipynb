Imputer

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
my_df2 = pd.DataFrame({"A" : [1,4,7,10,13],
                      "B" : [3,6,9,np.nan,15],
                      "C" : [2,5,np.nan,11,np.nan]})
my_df2
```

```python
#now we will call the class
imputer =SimpleImputer()
```

```python
#first we will apply fit and then do the tranform
imputer.fit(my_df2)
imputer.transform(my_df2)
```

```python
my_df2_1=imputer.transform(my_df2)
my_df2_1
```

```python
#we can apply fit and tranform at a time but not recommended
imputer.fit_transform(my_df2)
```

```python
#if u see previously it is in array so we converted to dataframe
my_df2_2=pd.DataFrame(imputer.fit_transform(my_df2), columns = ['A', 'B', 'C'])
my_df2_2
```
