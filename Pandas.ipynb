# Flitering

```python
import pandas as pd
```

```python
people = {
    "first": ["Corey", 'Jane', 'John'],
    "last": ["Schafer", 'Doe', 'Doe'],
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
df=pd.DataFrame(people)
df
```

```python
#the column name last matches with deo then it will return true
df['last']=='Doe'
```

```python
#another method using filt
filt=(df['last']=='Doe')
df[filt]
```

```python
#calling the filter with specified column
df.loc[filt,'email']
```

```python
#if both conditions satify
filt1=(df['last']=='Doe') &(df['first']=='John')
df.loc[filt1]
```

```python
#calling the filter with coulumn
df.loc[filt1,'email']
```

```python
#calling the filt1 which are opp to it
df.loc[~filt1,'email']
```

# cleaning data

```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'],
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df=pd.DataFrame(people)
df
```

```python
df.dropna()
```

```python
#any means if it has none/nan in the row then it will drop the row
df.dropna(axis='index',how='any')
```

```python
#if the whole row is nan /none only then it will drop
#u can see 4th row is missing
df.dropna(axis='index',how='all')
```

```python
#no whole columns has none values
#so no changes made
df.dropna(axis='columns',how='all')
```

```python
#u can see that each column has nan value soo the output is empty
df.dropna(axis='columns',how='any')
```

```python
# it will drop only one column that is email which contains nan none
df.dropna(axis='index',how ='any',subset=['email'])
```

```python
#it drop and gather the whole data either from last or email
df.dropna(axis='index',how ='all',subset=['last','email'])
```

```python
df.dropna(axis='index',how='any',subset=['last','email'])
```

```python
df
```

```python
df.replace('NA',np.nan,inplace=True)
df.replace('Missing',np.nan,inplace=True)
df
```
